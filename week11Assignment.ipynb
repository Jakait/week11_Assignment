```python
from sklearn.datasets import load_digits
     
```

```python
digits = load_digits()
```

```python
import matplotlib.pyplot as plt
```

```python
figure, axes = plt.subplots(nrows=4, ncols=6, figsize=(6, 4))
for item in zip(axes.ravel(), digits.images, digits.target):
    axes, image, target = item
    axes.imshow(image, cmap=plt.cm.gray_r)
    axes.set_xticks([])
    axes.set_yticks([])
    axes.set_title(target)
plt.tight_layout()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import numpy as np
import seaborn as sns
```

```python
knn = KNeighborsClassifier()
```

```python
#20% testing, 80% training
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)
```

```python
knn.fit(X=X_train, y=y_train)
```

```python
predicted = knn.predict(X=X_test)
expected = y_test
```

```python
wrong = [(p,e) for (p,e) in zip(predicted,expected) if p!=e]
```

```python
twentytest = knn.score(X=X_test, y=y_test)
twentytest
```

```python
confusiontwenty = confusion_matrix(y_true=expected, y_pred=predicted)
confusiontwenty
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(confusiontwenty, annot=True, fmt="d", cmap="Reds")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix for 20% Testing")
plt.show()
```

```python
total_predictions = np.sum(confusiontwenty)
total_predictions
```

```python
correct_predictions = np.trace(confusiontwenty)
correct_predictions
```

```python
#50% testing, 50% training
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.5, random_state=42)
knn.fit(X=X_train, y=y_train)
predicted = knn.predict(X=X_test)
expected = y_test
```

```python
fiftytest = knn.score(X=X_test, y=y_test)
fiftytest
```

```python
confusionfifty = confusion_matrix(y_true=expected, y_pred=predicted)
confusionfifty
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(confusionfifty, annot=True, fmt="d", cmap="Reds")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix for 50% Testing")
plt.show()
```

```python
#70% testing, 30% training
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.7, random_state=42)
knn.fit(X=X_train, y=y_train)
predicted = knn.predict(X=X_test)
expected = y_test
```

```python
seventytest = knn.score(X=X_test, y=y_test)
seventytest
```

```python
confusionseventy= confusion_matrix(y_true=expected, y_pred=predicted)
confusionseventy
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(confusionseventy, annot=True, fmt="d", cmap="Reds")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix for 70% Testing")
plt.show()
```
